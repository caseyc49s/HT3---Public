```python
# Convert UTCTime to Datetime
summary['UTCTime'] = pd.to_datetime(summary['UTCTime'].str.replace(",",""), unit="ms")

# Tried doing this the same way but ran into errors unless its an integer in days 
start_days = summary['StartTime'].str.replace(",","").astype(int)/(100*60*60*24)
end_days = summary['Endtime'].str.replace(",","").astype(int)/(100*60*60*24)

since_1900 = summary['TripStart'].str.replace(",","").astype(float)

days_delta = end_days - start_days
summary["days_delta"] = days_delta
summary["start_dt"] = pd.to_timedelta(start_days + since_1900, unit='d') + pd.datetime(1900,1,1)
summary["end_dt"] = pd.to_timedelta(days_delta, unit='d') + summary["start_dt"]

summary_dt = summary[["Device", "Trip", "start_dt", "end_dt"]]

summary.head()
```

```python
summary["days_deviation"] = summary["start_dt"].dt.day- summary["UTCTime"].dt.day
```

```python
summary.groupby(["days_deviation"]).nunique().loc[:, ["Device", "Trip"]]
```
